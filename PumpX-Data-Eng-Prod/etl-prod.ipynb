```python
import pandas as pd
import sqlalchemy, os
from dotenv import load_dotenv
import pymysql
from sqlalchemy import create_engine, text
```

```python
load_dotenv('config.env')
username = os.getenv("DB_USERNAME")
password = os.getenv("DB_PASSWORD")
host = os.getenv("DB_HOST")
database_name = os.getenv("DB_NAME")

connect = f"mysql+pymysql://{username}:{password}@{host}/{database_name}"
engine = sqlalchemy.create_engine(connect)
conn = engine.connect()
```

```python
def transform_user_data():
    q = text(
        '''
        SELECT *
        FROM user
        WHERE email NOT IN (
            SELECT DISTINCT SUBSTRING(email, 1, IF(LOCATE('+', email) > 0, LOCATE('+', email) - 1, LOCATE('@', email) - 1)) AS prefix
            FROM user
            WHERE (email REGEXP '^[a-zA-Z]+\\.[a-zA-Z]+\\d*@gmail\\.com$' OR email REGEXP '^[a-zA-Z]+\\.[a-zA-Z]+\\+\\d*@gmail\\.com$')
            GROUP BY prefix
            HAVING COUNT(*) > 1
        ) AND NOT (email LIKE '%test%' OR username LIKE '%test%');
    '''
    )
    ex = conn.execute(q)
    return ex.fetchall()

for a in transform_user_data():
    print(a)



def get_total_loggings_week():
    q = '''
        SELECT DATE(approach_time) AS logging_date, COUNT(DISTINCT user_id) AS number_users 
        FROM pumpxdb_prod.exercise_set 
        WHERE YEARWEEK(DATE(approach_time), 1) = YEARWEEK(CURDATE(), 1) GROUP BY logging_date;
    '''
```
